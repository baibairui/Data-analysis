```python
#将数据分为数值型和非数值型
data_num=data.select_dtypes(exclude=['object'])
data_obj=data.select_dtypes(include=['object'])
data_num.head()
```

```python
#数值型数据的缺失值处理
data_num=data_num.fillna(data_num.mean())
data_num.head()
```

```python
#取出房价和年份月份
train_data=data_num[['SalePrice','MoSold','YrSold']]
train_data.head()
```

```python
#将年份和月份合并
train_data['time']=train_data['YrSold'].astype(str)+'-'+train_data['MoSold'].astype(str)
train_data.head()
```

```python
#将时间转换为时间戳
train_data['time']=pd.to_datetime(train_data['time'])
#去除年份和月份
train_data=train_data.drop(['YrSold','MoSold'],axis=1)
train_data.head()
```

```python
#数据平滑化
smoothed_data = train_data['SalePrice'].rolling(window=7, min_periods=1).mean()
#查看数据
smoothed_data.head()
```

```python
#将平滑化后的数据和原数据合并
train_data['SalePrice']=smoothed_data
train_data.head()
```

```python
#打印acf和pacf
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_acf(train_data['SalePrice'])
plot_pacf(train_data['SalePrice'])
plt.show()
```

```python
#构建arima模型

# Fit the model自动选择最优的p,q,d
#使用 pmdarima 自动寻找最佳 ARIMA 模型参数
import pmdarima as pm

model = pm.auto_arima(train_data.iloc[:,1], seasonal=False)

# 打印模型摘要
print(model.summary())
```

```python
#构建模型
from statsmodels.tsa.arima.model import ARIMA

# Re-fitting the ARIMA model to the smoothed data
model_smoothed = ARIMA(train_data['SalePrice'], order=(1,0,1))
model_fit_smoothed = model_smoothed.fit()
#查看模型的参数
model_fit_smoothed.summary()
```

```python
#预测
# Forecasting
# Plotting the forecasted values
plt.figure(figsize=(20,10))#设置画布大小，第一个参数是宽度，第二个参数是高度
plt.plot(train_data['SalePrice'], label='Actual')
plt.plot(model_fit_smoothed.fittedvalues, color='red', label='Predicted')
plt.legend()
plt.show()
```

```python
#预测未来的房价
# Forecasting the future
# Predicting the future values
future = model_fit_smoothed.predict(start=len(train_data)+1, end=len(train_data)+1459)
future 
```
