```python
#读取数据
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_excel('data.xlsx')#从第一行开始读取数据
data.head()#显示前五行数据
```

```python
#数据预处理
#去除第一列
data=data.iloc[:,1:]
data.head()
```

```python
data=data.values
data=pd.DataFrame(data)
data.head()
```

```python
#检测第三列中字母数不为5的数据
data_error=data[data.iloc[:,2].str.len()!=5]   #检测第三列中字母数不为5的数据
data_error
```

```python
#检测5-到11列之和不为1的数据
data_error2=data[data.iloc[:,5:11].sum(axis=1)!=1]   #检测5-到11列之和不为1的数据
data_error2.head()
```

```python
#将异常数据合并后导入到excel
data_error=pd.concat([data_error,data_error2])
data_error.to_excel('data_error.xlsx')
data[0:10]
```
